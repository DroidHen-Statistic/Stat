```python
import pymysql
import pymysql.cursors

class MysqlConnection(object):
    def __init__(self,host,user,password,db,charset = 'utf8',cursorclass = pymysql.cursors.DictCursor):
        self.__connection = pymysql.connect(host=host,user=user,password=password,db=db,charset=charset,cursorclass = cursorclass)

    def close(self):
        self.__connection.close()

    def query(self,sql,values=()):
        result = []
        #获取一个游标
        with self.__connection.cursor() as cursor:
            cout = cursor.execute(sql,values)
            self.__connection.commit()
            for row in cursor.fetchall():
                #result.append(list(row))
                result.append(row)
                #print('%s\t%s\t%s' %row
        return result
```

## Visualization

```python
# first ten features
y_tmp = y
x_tmp = x
x_n_2 = (x_tmp - x_tmp.mean()) / (x_tmp.std())              # standardization
df = pd.concat([y_tmp,x_n_2.iloc[:,0:10]],axis=1)
df = pd.melt(df,id_vars="purchase_times",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", hue="purchase_times", data=df,split=True, inner="quart")
plt.xticks(rotation=90)
plt.show()
```

```python
# first ten features
y_tmp = y
x_tmp = x
x_n_2 = (x_tmp - x_tmp.mean()) / (x_tmp.std())              # standardization
df = pd.concat([y_tmp,x_n_2.iloc[:,10:20]],axis=1)
df = pd.melt(df,id_vars="purchase_times",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", hue="purchase_times", data=df,split=True, inner="quart")
plt.xticks(rotation=90)
plt.show()
```
